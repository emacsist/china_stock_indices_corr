```python
%matplotlib inline

import matplotlib as mpl 
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
import seaborn as sns
from functools import reduce

plt.rcParams['font.sans-serif'] = ['Arial Unicode MS'] 
mpl.rcParams["axes.unicode_minus"] = False
```

```python
df000001 =pd.read_csv('000001.csv.ok', parse_dates=True, index_col='日期')[['名称', '收盘价']]
df000016 =pd.read_csv('000016.csv.ok', parse_dates=True, index_col='日期')[['名称', '收盘价']]
df000903 =pd.read_csv('000903.csv.ok', parse_dates=True, index_col='日期')[['名称', '收盘价']]
df000919 =pd.read_csv('000919.csv.ok', parse_dates=True, index_col='日期')[['名称', '收盘价']]
df000922 =pd.read_csv('000922.csv.ok', parse_dates=True, index_col='日期')[['名称', '收盘价']]
df000932 =pd.read_csv('000932.csv.ok', parse_dates=True, index_col='日期')[['名称', '收盘价']]
df000978 =pd.read_csv('000978.csv.ok', parse_dates=True, index_col='日期')[['名称', '收盘价']]
df399001 =pd.read_csv('399001.csv.ok', parse_dates=True, index_col='日期')[['名称', '收盘价']]
df399006 =pd.read_csv('399006.csv.ok', parse_dates=True, index_col='日期')[['名称', '收盘价']]
df399300 =pd.read_csv('399300.csv.ok', parse_dates=True, index_col='日期')[['名称', '收盘价']]
df399324 =pd.read_csv('399324.csv.ok', parse_dates=True, index_col='日期')[['名称', '收盘价']]
df399905 =pd.read_csv('399905.csv.ok', parse_dates=True, index_col='日期')[['名称', '收盘价']]
df399975 =pd.read_csv('399975.csv.ok', parse_dates=True, index_col='日期')[['名称', '收盘价']]
df399997 =pd.read_csv('399997.csv.ok', parse_dates=True, index_col='日期')[['名称', '收盘价']]
dfs = [df000001,df000016,df000903,df000919,df000922,df000932,df000978,df399001,df399006,df399300,df399324,df399905,df399975,df399997]

df_final = reduce(lambda left,right: pd.merge(left,right,left_index=True, right_index=True), dfs)
```
